```python
import pandas as pd

# Load the dataset
file_path = 'house_sales.csv'  # Adjust the path as needed
df = pd.read_csv(file_path)
df
```

```python
null_values = df.isnull().sum()

duplicates = df.duplicated().sum()

df = df.drop_duplicates()

df = df.dropna()

print("Null values per column:\n", null_values)
print("Number of duplicate rows:", duplicates)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['price'] = df['price'].astype(float)
df['bedrooms'] = df['bedrooms'].astype(int)

# print("Data types before conversion:\n", data_types)

# data_types_after_conversion = df.dtypes
# print("Data types after conversion:\n", data_types_after_conversion)

df.dtypes
```

```python
import matplotlib.pyplot as plt

# Select numeric columns
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms for each numeric column
df[numeric_columns].hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
sample_data = df.head()

summary_data = df.describe()

sample_data
summary_data
```

```python
first_sale_date = df['date'].min()
last_sale_date = df['date'].max()

print("First sale date:", first_sale_date)
print("Last sale date:", last_sale_date)

# Plot the distribution of sales over time
df['date'].value_counts().sort_index().plot(kind='line', figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Number of Sales')
plt.title('Distribution of Sales Over Time')
plt.show()
```

```python
# Extract relevant time periods
df['quarter'] = df['date'].dt.to_period('Q')
df['month'] = df['date'].dt.to_period('M')
df['week'] = df['date'].dt.to_period('W')
df['day'] = df['date'].dt.date

# Group by quarter, month, week, and day
quarter_sales = df.groupby('quarter')['price'].sum()
month_sales = df.groupby('month')['price'].sum()
week_sales = df.groupby('week')['price'].sum()
day_sales = df.groupby('day')['price'].sum()

# Plot sales amounts by quarter
quarter_sales.plot(kind='line', figsize=(12, 6), title='Total Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Amount')
plt.show()

# Plot sales amounts by month
month_sales.plot(kind='line', figsize=(12, 6), title='Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.show()

# Plot sales amounts by week
week_sales.plot(kind='line', figsize=(12, 6), title='Total Sales by Week')
plt.xlabel('Week')
plt.ylabel('Total Sales Amount')
plt.show()

# Plot sales amounts by day
day_sales.plot(kind='line', figsize=(12, 6), title='Total Sales by Day')
plt.xlabel('Day')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
zipcode_stats = df.groupby('zipcode').agg(
    houses_sold=('id', 'count'),
    avg_condition=('condition', 'mean'),
    total_revenue=('price', 'sum'),
    first_year_built=('yr_built', 'min'),
    last_year_built=('yr_built', 'max')
).reset_index()

zipcode_stats
```

```python
# Plot distribution of houses sold by zipcode
zipcode_stats['houses_sold'].plot(kind='hist', bins=30, figsize=(12, 6), title='Distribution of Houses Sold by Zipcode')
plt.xlabel('Number of Houses Sold')
plt.show()

# Plot distribution of average condition by zipcode
zipcode_stats['avg_condition'].plot(kind='hist', bins=30, figsize=(12, 6), title='Distribution of Average Condition by Zipcode')
plt.xlabel('Average Condition')
plt.show()

# Plot distribution of total revenue by zipcode
zipcode_stats['total_revenue'].plot(kind='hist', bins=30, figsize=(12, 6), title='Distribution of Total Revenue by Zipcode')
plt.xlabel('Total Revenue')
plt.show()

# Plot distribution of first year built by zipcode
zipcode_stats['first_year_built'].plot(kind='hist', bins=30, figsize=(12, 6), title='Distribution of First Year Built by Zipcode')
plt.xlabel('First Year Built')
plt.show()

# Plot distribution of last year built by zipcode
zipcode_stats['last_year_built'].plot(kind='hist', bins=30, figsize=(12, 6), title='Distribution of Last Year Built by Zipcode')
plt.xlabel('Last Year Built')
plt.show()
```

```python
zipcode_most_sales = zipcode_stats.loc[zipcode_stats['houses_sold'].idxmax()]
zipcode_highest_revenue = zipcode_stats.loc[zipcode_stats['total_revenue'].idxmax()]

zipcode_most_sales
zipcode_highest_revenue
```

```python
top_100_houses = df.nlargest(100, 'price')

top_100_houses
```

```python
numeric_columns_top_100 = top_100_houses.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms for each numeric column
top_100_houses[numeric_columns_top_100].hist(bins=30, figsize=(15, 10))
plt.suptitle('Distribution of Numeric Attributes for Top 100 Houses by Sale Price', y=1.02)
plt.tight_layout()
plt.show()
```

```python
top_100_houses[numeric_columns_top_100].plot(kind='box', subplots=True, layout=(5, 5), figsize=(15, 10), sharex=False, sharey=False)
plt.suptitle('Boxplots of Numeric Attributes for Top 100 Houses by Sale Price', y=1.02)
plt.tight_layout()
plt.show()
```

```python
bedroom_groups = top_100_houses['bedrooms'].value_counts().sort_index()

# Plot the number of houses in each bedroom group
bedroom_groups.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Number of Bedrooms')
plt.ylabel('Number of Houses')
plt.title('Distribution of Top 100 Houses by Number of Bedrooms')
plt.show()
```

```python
top_100_houses.boxplot(column='price', by='bedrooms', figsize=(10, 6))
plt.xlabel('Number of Bedrooms')
plt.ylabel('Sale Price')
plt.title('Boxplot of Sale Prices for Top 100 Houses Grouped by Number of Bedrooms')
plt.suptitle('')  # Suppress the default title to keep the plot clean
plt.show()
```
