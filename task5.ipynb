```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'house_sales.csv'  # Adjust the path as needed
df = pd.read_csv(file_path)

# Compute the correlation matrix
df['date'] = pd.to_datetime(df['date'])
df['price'] = df['price'].astype(float)
df['id'] = df['id'].astype(str)

null_values = df.isnull().sum()

duplicates = df.duplicated().sum()

df = df.drop_duplicates()

df = df.dropna()

correlation_matrix = df.corr()

correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns
relevant_columns = ['price', 'sqft_living', 'sqft_above', 'grade']

# Plot the pairwise correlation using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df[relevant_columns].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pairwise Correlation of Selected Features')
plt.show()
```

```python
# Scatter plot for number of bedrooms vs. house price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bedrooms', y='price')
plt.title('Impact of Number of Bedrooms on House Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('House Price')
plt.show()

# Scatter plot for number of bathrooms vs. house price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bathrooms', y='price')
plt.title('Impact of Number of Bathrooms on House Price')
plt.xlabel('Number of Bathrooms')
plt.ylabel('House Price')
plt.show()

# Scatter plot for year of construction vs. house price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='yr_built', y='price')
plt.title('Impact of Year of Construction on House Price')
plt.xlabel('Year of Construction')
plt.ylabel('House Price')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Standardize the dataset
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df.select_dtypes(include=['float64', 'int64'])), columns=df.select_dtypes(include=['float64', 'int64']).columns)

# Select the feature with the highest correlation to house price
correlation_with_price = correlation_matrix['price'].drop('price')
best_feature = correlation_with_price.idxmax()

# Fit a linear regression model
X = df_standardized[[best_feature]]
y = df_standardized['price']

model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plot the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[best_feature], y=y, label='Actual')
sns.lineplot(x=X[best_feature], y=y_pred, color='red', label='Predicted')
plt.xlabel(best_feature)
plt.ylabel('Standardized Price')
plt.title(f'Linear Regression: {best_feature} vs. House Price')
plt.legend()
plt.show()

# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
